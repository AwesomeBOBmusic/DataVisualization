# ***Interquartile***

```python
states_df = pd.DataFrame(data = [
['NJ', 'Towaco', 25, 1, 'Sunny'],
['CA', 'San Francisco', 40, 0, 'Sunny'],
['TX', 'Austin', -100, 0, 'Cloudy'],
['MD', 'Baltimore',21, 1, 'Cloudy'],
['OH', 'Columbus', 31, 1, 'Sunny'],
['IL', 'Chicago', 16, 0, 'Cloudy']
],
  columns = ['State', 'City', 'Temp', 'Happy', 'Conditions']
            )
```

```python
import pandas as pd

Q1 = states_df['Temp'].quantile(0.25)
Q3 = states_df['Temp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = states_df[(states_df['Temp'] < lower_bound) | (states_df['Temp'] > upper_bound)]

outliers
```

```python
states_df_cop = states_df.copy()

Q1 = states_df_cop.quantile(0.25, numeric_only=True)
Q3 = states_df_cop.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

# ***Scaling***

```python
# For min_max scaling
from sklearn.preprocessing import MinMaxScaler

# Select the temp column
temp_original = states_df.Temp

# Scale the goals from 0 to 1 using MinMaxScaler()
# fit computes the minimum and maximum to be used for later scaling, and transform performs the scaling
temp_scaled = MinMaxScaler().fit_transform(states_df[["Temp"]])

# Plot the original & scaled data together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(temp_original, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(temp_scaled, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
from sklearn.preprocessing import StandardScaler

temp_z_scaled = StandardScaler().fit_transform(states_df[["Temp"]])

# Plot the original & scaled data with both methods to compare
fig, ax=plt.subplots(1,3, constrained_layout=True) # layout spaces out the figures for more visibility
sns.distplot(temp_original, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(temp_scaled, ax=ax[1])
ax[1].set_title("Scaled data - min-max")
sns.distplot(temp_z_scaled, ax=ax[2])
ax[2].set_title("Scaled data - z-score")
```

```python
# get only positive pledges (using their indexes)
positive_pledges = states_df.Temp.loc[states_df.Temp > 0]

# normalize the pledges (w/ Box-Cox)
normalized_pledges = stats.boxcox(positive_pledges)[0]
# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_pledges, ax=ax[1])
ax[1].set_title("Normalized data")
```

# ***Label Encoding***

```python
import sklearn.preprocessing as preprocessing

# Function to encode categorical attributes in a given dataset as numbers
def number_encode_features(df):
    result = states_df.copy() # take a copy of the dataframe
    for column in result.columns:
        if result.dtypes[column] == np.object_: # if attribute is categorical
            # Apply LabelEncoder method to attribute
            # fit will infer the number of numerical values needed by counting the number of categories
            # then transform will replace each category with its numerical counterpart
            result[column] = preprocessing.LabelEncoder().fit_transform(result[column])
    return result

# Apply function defined above to income dataset
encoded_data = number_encode_features(states_df)
# Now that the attributes are numerical, it is possible to calculate correlation and plot it
sns.heatmap(encoded_data.corr(), square=True, cmap='PuRd')
plt.show()
```

# ***Discretization***

```python
states_df_dis = pd.DataFrame(data = [
['NJ', 'Towaco', 25, 1, 'Sunny'],
['CA', 'San Francisco', 40, 0, 'Sunny'],
['TX', 'Austin', 40, 0, 'Cloudy'],
['MD', 'Baltimore',4, 1, 'Cloudy'],
['OH', 'Columbus', 31, 1, 'Sunny'],
['IL', 'Chicago', 16, 0, 'Cloudy']
],
  columns = ['State', 'City', 'Temp', 'Happy', 'Conditions']
            )
```

```python
temp_group = []
for temps in states_df_dis["Temp"]:
    if temps < 5:
        temp_group.append("<5")
    elif 5 <= temps <= 15:
        temp_group.append("5-15")
    elif 15 < temps <= 25:
        temp_group.append("15-25")
    elif 35 < temps <= 45:
        temp_group.append("35-45")
    else:
        temp_group.append("45 and over")

states_df_dis["Temp_group"] = temp_group
del states_df_dis["Temp"]
states_df_dis
```

```python
# Visualize the discretized data
plt.figure(figsize=(8, 6))
states_df_dis['Temp_group'].value_counts(sort=False).plot(kind='bar')
plt.xlabel('Temp Range')
plt.ylabel('Count in Group')
plt.title('Number of values in a Temperature group')
plt.show()
```

```python
# Removed the temp column
states_df_similar = pd.DataFrame(data = [
['NJ', 'Sunny', 'Sunny'],
['CA', 'Sunny', 'Sunny'],
['TX', 'Cloudy', 'Cloudy'],
['MD', 'Cloudy', 'Cloudy'],
['OH', 'Sunny', 'Sunny'],
['IL', 'Cloudy', 'Cloudy']
],
  columns = ['State', 'Potato', 'Conditions']
            )
```

```python
states_df_similar.head()
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(states_df_similar['Potato'], states_df_similar['Conditions'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the results
alpha = 0.05
if p < alpha:
    print("There is a statistically significant correlation between Happy and Condition.")
else:
    print("There is no statistically significant correlation between Happy and Condition. They appear to be independent.")
```

```python
import pyspark
from pyspark import SparkFiles
from pyspark.sql import SparkSession
from pyspark import SparkContext
spark = SparkSession.builder.appName("word count").getOrCreate()
```

```python
# Outputs each word on each line
url = get_path("word_count.text")
spark.sparkContext.addFile(url)
lines = spark.sparkContext.textFile("file://"+SparkFiles.get("word_count.text"))
x1 = lines.flatMap(lambda line: line.split(" "))
for w in x1.collect() :
  print(w)
```

```python
# Counts each word
counts = lines.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
```

```python
output = counts.collect() # Returns a list with all the rows (counts)
for (word, count) in output:
    print("%s: %i" % (word, count))
```
